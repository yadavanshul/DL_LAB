```python
anime_df['Duration'] = anime_df['Duration'].replace('Unknown', np.nan)
```

```python
import re
def extract_duration(duration):
    if pd.isnull(duration):
        return np.nan
    # Extract hours and minutes
    hr_match = re.search(r'(\d+)\s*hr', duration)
    min_match = re.search(r'(\d+)\s*min', duration)
    
    hours = int(hr_match.group(1)) * 60 if hr_match else 0
    minutes = int(min_match.group(1)) if min_match else 0
    
    return hours + minutes

# Apply the extraction function to the 'Duration' column
anime_df['Duration_in_min'] = anime_df['Duration'].apply(extract_duration)

# Step 2: Replace NaN with the mean of the 'Duration_in_min' column
mean_duration = anime_df['Duration_in_min'].mean()
anime_df['Duration_in_min'].fillna(mean_duration, inplace=True)
```

```python
anime_df.drop(columns=['Name', 'English name', 'Japanese name','sypnopsis','Aired', 'Premiered','Producers','Licensors','Studios','Genres','Duration'], inplace=True)
anime_df
```

```python
anime_df['Episodes'] = anime_df['Episodes'].replace('UNKNOWN', 0)

# Convert the 'Episodes' column to numeric (this will convert it from object type to numerical type)
anime_df['Episodes'] = pd.to_numeric(anime_df['Episodes'], errors='coerce')

# Calculate the mean of the 'Episodes' column, ignoring zero values
mean_episodes = anime_df[anime_df['Episodes'] != 0]['Episodes'].mean()

# Replace 0 values with the mean of the 'Episodes' column
anime_df['Episodes'] = anime_df['Episodes'].replace(0, mean_episodes)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

# Define a custom dataset
class AnimeDataset(Dataset):
    def __init__(self, data):
        self.data = data
        # Separate categorical and numerical data
        self.user_ids = torch.tensor(data['user_id'].values, dtype=torch.long)
        self.anime_ids = torch.tensor(data['anime_id'].values, dtype=torch.long)
        self.numerical_features = torch.tensor(data[numerical_features].values, dtype=torch.float)
        self.labels = torch.tensor(data['rating'].values, dtype=torch.float)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return (self.user_ids[idx], self.anime_ids[idx], self.numerical_features[idx]), self.labels[idx]

# Wide and Deep Model Definition
class WideAndDeepModel(nn.Module):
    def __init__(self, num_users, num_anime, embedding_dim=32, num_numerical_features=8):
        super(WideAndDeepModel, self).__init__()
        # Embeddings for categorical features
        self.user_embedding = nn.Embedding(num_users, embedding_dim)
        self.anime_embedding = nn.Embedding(num_anime, embedding_dim)
        # Update total input dimension after concatenation to match observed shape
        total_input_dim = 73
        # Fully connected layers for wide component
        self.wide_fc1 = nn.Linear(total_input_dim, 128)
        self.wide_dropout = nn.Dropout(0.2)
        # Fully connected layers for deep component
        self.deep_fc1 = nn.Linear(total_input_dim, 256)
        self.deep_dropout1 = nn.Dropout(0.2)
        self.deep_fc2 = nn.Linear(256, 128)
        self.deep_dropout2 = nn.Dropout(0.2)
        self.deep_fc3 = nn.Linear(128, 64)

        # Final output layer
        self.output_layer = nn.Linear(128 + 64, 1)

    def forward(self, user_id, anime_id, numerical_features):
        # Embedding lookups
        user_embedded = self.user_embedding(user_id)
        anime_embedded = self.anime_embedding(anime_id)
        
        # Concatenate embeddings and numerical features
        x = torch.cat([user_embedded, anime_embedded, numerical_features], dim=1)
        # Wide part
        wide_out = self.wide_fc1(x)
        wide_out = self.wide_dropout(wide_out)
        wide_out = torch.relu(wide_out)
        # Deep part
        deep_out = torch.relu(self.deep_fc1(x))
        deep_out = self.deep_dropout1(deep_out)
        deep_out = torch.relu(self.deep_fc2(deep_out))
        deep_out = self.deep_dropout2(deep_out)
        deep_out = torch.relu(self.deep_fc3(deep_out))
        # Concatenate wide and deep parts
        combined = torch.cat([wide_out, deep_out], dim=1)# Final output layer without sigmoid for regression
        output = self.output_layer(combined)
        return output


# Training and Evaluation Functions
def train_model(model, dataloader, optimizer, criterion, epochs=5):
    model.train()
    for epoch in range(epochs):
        total_loss = 0
        i = 0
        for (user_id, anime_id, numerical_features), labels in dataloader:
            i += 1
            optimizer.zero_grad()
            outputs = model(user_id, anime_id, numerical_features)
            loss = criterion(outputs.squeeze(), labels)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
            
            if (i + 1) % 200 == 0:
                print(f'Step size {i + 1}, Loss: {loss.item()}')
        evaluate_model(model, dataloader)  
        print(f'Epoch {epoch + 1}, Loss: {total_loss / len(dataloader)}')
        torch.save(model.state_dict(), f'model_epoch_{epoch + 1}.pth')


def evaluate_model(model, dataloader):
    model.eval()
    correct, total = 0, 0
    with torch.no_grad():
        for (user_id, anime_id, numerical_features), labels in dataloader:
            outputs = model(user_id, anime_id, numerical_features)
            predicted = (outputs.squeeze() > 0.5).float()
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    accuracy = correct / total
    print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
# Model Parameters and Initialization
# Reindexing user_id and anime_id to a continuous range
user_id_mapping = {id: idx for idx, id in enumerate(df['user_id'].unique())}
anime_id_mapping = {id: idx for idx, id in enumerate(df['anime_id'].unique())}

# Apply mappings
df['user_id'] = df['user_id'].map(user_id_mapping)
df['anime_id'] = df['anime_id'].map(anime_id_mapping)

num_users = len(df['user_id'].unique())
num_anime = len(df['anime_id'].unique())
numerical_features = ['Days Watched', 'Mean Score', 'Ranked', 'Popularity', 'Members', 'Favorites', 'Duration_in_min','Gender','Rewatched']
dataset = AnimeDataset(df)
dataloader = DataLoader(dataset, batch_size=1048, shuffle=True)

model = WideAndDeepModel(num_users=num_users, num_anime=num_anime)
optimizer = optim.AdamW(model.parameters(), lr=0.001)
criterion = nn.SmoothL1Loss()
```
